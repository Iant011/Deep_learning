```python
# Required Lab
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.read_csv('https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv')
```

```python
df.head()
```

```python
## Converting dataset into dependants and independants
X = df.iloc[:,3:13]
y = df.iloc[:,13]
```

```python
X.head()
```

```python
y.head()
```

```python
## Converting categorical values into Onehot Encoding

geography = pd.get_dummies(X['Geography'])
gender = pd.get_dummies(X['Gender'])
```

```python
X = X.drop(['Geography','Gender'],axis=1)
```

```python
X.head()
```

```python
X = pd.concat([X,geography,gender],axis = 1)
```

```python
X.head()
```

```python
# Converting data into train test split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler   ## Based on Z_score test

sc = StandardScaler()

X_train = sc.fit_transform(X_train)

X_test = sc.transform(X_test)
```

```python
# ANN
```

```python
from tensorflow.keras.models import Sequential #FOrward backward --> epochs
from tensorflow.keras.layers import Dense  # Creating circles layers
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU #--> Activation Functions
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units = 11,activation = 'relu'))
```

```python
classifier.add(Dense(units = 7,activation = 'relu'))
```

```python
classifier.add(Dense(units = 5,activation = 'relu'))
```

```python
# Output layers
classifier.add(Dense(1,activation='sigmoid'))
```

```python
from keras.src import optimizers
from keras.api._v2.keras import metrics

classifier.compile(optimizer = 'adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Learning rate
import tensorflow
tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
from sklearn.utils import validation
model_history = classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=1000)
```
